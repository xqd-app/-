# 数据探索与分析

在本Notebook中，我们将进行数据的探索和分析，以便更好地理解社交关系网络的特征。

```python
# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置Seaborn样式
sns.set(style='whitegrid')
```

```python
# 加载数据
data = pd.read_csv('../data/processed/social_network_data.csv')

# 显示数据的前五行
data.head()
```

```python
# 数据基本信息
data.info()
```

```python
# 数据描述统计
data.describe()
```

```python
# 可视化社交网络的度分布
plt.figure(figsize=(10, 6))
sns.histplot(data['degree'], bins=30, kde=True)
plt.title('社交网络的度分布')
plt.xlabel('度')
plt.ylabel('频率')
plt.show()
```
